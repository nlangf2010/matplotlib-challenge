## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_db = pd.merge(left=mouse_metadata, right=study_results, left_on='Mouse ID', right_on='Mouse ID')

# Display the data table for preview
merged_db.head(100)


```

```python
# Checking the number of mice.
merged_db["Mouse ID"].count()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
merged_db.duplicated().sum()
merged_db.loc[merged_db.duplicated(keep = False), :]


```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_db = merged_db.drop_duplicates()
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_db['Mouse ID'].count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

cleaned_db['Tumor Volume (mm3)'].describe()

# This method is the most straighforward, creating multiple series and putting them all together at the end.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
tumor_vol_data = cleaned_db.groupby(['Drug Regimen', 'Mouse ID'])
tumor_vol_data.count().head(40)



# This method produces everything in a single groupby function
tumor_vol_df = pd.DataFrame(
    tumor_vol_data["Tumor Volume (mm3)"].describe().dropna().reset_index())

tumor_vol_df.head(40)
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 

treatment_and_mice = cleaned_db.groupby('Drug Regimen')

treatment_df = treatment_and_mice['Drug Regimen'].count()

treatment_df.plot(x='Drug Regimen', kind='bar')






```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.

# Create a group based on the values in the 'Drug Regimen' column
treatment_group = cleaned_db.groupby('Drug Regimen')

# Count how many times each Drug Regimen appears in our group
count_treatment = treatment_group['Drug Regimen'].count()

count_treatment

# Create a bar chart based off of the group series from before
count_chart = count_treatment.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Treatment Name")
count_chart.set_ylabel("Number of Mice")

plt.title("Number of Mice For Each Treatment Throughout Study")
plt.show()
plt.tight_layout()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
mice_gender = cleaned_db.groupby('Sex')

mice_gender_dist = mice_gender['Sex'].count()
mice_gender_dist

mice_gender_dist.plot(kind='pie')
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
gender_pie_list = mice_gender_dist.keys()

gender_pie = mice_gender_dist.plot(kind="pie", y=gender_pie_list, title=("Distribution of Female vs. Male Mice"))

gender_pie.set_ylabel("Gender Distribution")

plt.show()
plt.tight_layout()
plt.axis("equal")


```
